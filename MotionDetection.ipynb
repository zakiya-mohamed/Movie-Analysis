## Motion Detection
Program that detects motion between two consecutive frames in a video (`.mp4`) file. Frames are read as arrays of pixels. Each pixel array contains pixel values in BGR format (i.e., each pixel array is 3-dimensional).

### Method and assumptions
If the **average absolute difference** (obtained by calculating the absolute difference between two multi-dimensional arrays of pixels) between two consecutive frames is **greater than 55** (***NOTE***: *this value is obtained by experimentation and will vary depending on the input video/image files*), then **significant motion** has occured, otherwise, we assume **no significant motion** has occured, and set the average difference between frames as **zero**.

A threshold of 55 for the average difference between two consecutive frames is selected in order to clearly display sections of significant motion **only**

We utilize the python module `opencv` to compute the **absolute difference** between two consecutive movie frames.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

3) Capture movie

```python
# Capture video
cap = cv2.VideoCapture('Logan_2017.mp4')
```

4) Count total number of frames in video

```python
# Total number of frames in video 
length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f'Total number of frames in video: {length}')
```

5) Read first two frames of video 

```python
# Read two frames, last and current
ret, last_frame = cap.read()
ret, current_frame = cap.read()
```

6) Find difference between consecutive frames in video for all frames

```python
i = 0
absdiff = []
while(True):
  # We want two frames, last and current, so that we can calculate the difference between them
  # Store the current frame as last_frame, and then read a new one
  last_frame = current_frame

  ret, current_frame = cap.read()

  # Find the absolute difference between frames
  diff = cv2.absdiff(last_frame, current_frame)
  avgdiff = np.mean(diff)

  # If average difference between frames > 55, motion detected
  # Otherwise, no motion detected 
  if avgdiff > 55:
    absdiff.append(avgdiff)
  else:
    absdiff.append(0)
    print('No motion detected')

  # Print output every 10 frames
  i += 1
  if i % 10 == 0:
    i = 0
    print(f'Current frame: {np.mean(current_frame)} ---> Difference: {avgdiff}')
```

7) Check length of list of differences

```python
len(absdiff)
```

8) Plot differences (i.e., movement detection) between frames

```python
plt.figure(figsize = (15, 10))
ax = plt.axes()
ax.set_facecolor('w')
plt.plot(absdiff, linewidth = 0.4, alpha = 0.8, color = 'brown')
plt.title('Logan (2017)')
plt.axis('off')
plt.show()
```
