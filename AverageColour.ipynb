## Average Colour of a Frame

Program that uses the K-Means Algorithm to generate the **average** colour of a particular input frame. This is done by setting the number of K-Mean clusters to 1 (i.e., one centroid **only**)

```python
from sklearn.cluster import KMeans
from collections import Counter
from pathlib import Path

import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

```python
# Change working directory 
import os 
os.chdir('/Users/nmfamily/Desktop')
os.getcwd() 
```

1) Process and display specified movie still

```python
# Specify path to movie still and display still
still_path = Path("Logan_2017_2.png")
still = Image.open(still_path.open("rb"))
still
```

2) Reshape movie still into a list of pixels

```python
arr = np.array(still)
pix = arr.reshape((-1,4))
```

3) Center pixels between values of 0 and 1 

```python
pix = pix.astype("float32") / 255
```

4) Apply K-Means clustering with one cluster/centroid **ONLY**

```python
kmeans = KMeans(n_clusters = 1)  # Specify K-Means clustering with 1 centre
preds = kmeans.fit_predict(pix)
centers = kmeans.cluster_centers_
```

5) Plot average colour of input movie still

```python
plt.figure(figsize = (14,8))
plt.imshow(centers[np.concatenate([[i] * 100 for i in range(len(centers))]).reshape((-1, 10)).T])
plt.title("Logan 2 (2017) Frame 2 - Average Color")
plt.show()
```
