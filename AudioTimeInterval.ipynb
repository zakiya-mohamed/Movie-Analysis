## Finding the ideal time interval for audio analysis
Since *Ad Astra (2019)* seems to be the movie with the most 'silent' sections in the audio waveform compared to the remaining movies, we decided to experiment with various time intervals in order to select which time interval is ideal to obtain RMS values, and be able to see clear differences between them in a plot

### Overall method
We experiment with the following time intervals: **5 seconds, 10 seconds, 15 seconds, 20 seconds, 25 seconds, and 30 seconds** (i.e., 6 different time intervals) and determine which is best for this particular analysis

### What next?
Once we decide whic time interval is ideal for this particular movie, we use the same interval for the remaining movies for uniformity 

> *NOTE*: For this particular movie, we conclude that a **15 second time 
interval** is ideal

```python
from pydub import AudioSegment
import matplotlib.pyplot as plt
import librosa
import numpy as np
%matplotlib inline
```

2) Read mp3 file 

```python
# Read .mp3 file
mp3 = AudioSegment.from_file('AA_2019.mp3', format = 'mp3')
```

3) Split sound into **5 second time intervals** and obtain RMS at each interval

```python
# Split sound into T second slices and export
T = 5          # T second interval
aud_rms = []
datapt = []
for i, chunk in enumerate(mp3[::T*1000]):
  sound = chunk.rms
  aud_rms.append(sound)
  datapt.append(i)
```

4) Check number of data points

```python
len(aud_rms)
```

5) Plot RMS 

```python
datapt_rescaled = [i * T/60 for i in datapt]
fig = plt.figure(figsize = (15, 10))
fig.patch.set_facecolor('w')
plt.plot(datapt_rescaled, aud_rms, color = 'darkorange', linewidth = 1.5)
plt.title('Ad Astra (2019)')
plt.axis('off')
plt.show()
```
