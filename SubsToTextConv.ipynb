4) Remove additional unwanted characters from output `.txt` file and export as final `.txt` file

```python
with open(r'Logan_2017.txt', 'r') as infile, \
     open(r'NEW_Logan_2017.txt', 'w') as outfile:
    data = infile.read()
    data = data.replace("<i>", "")
    data1 = data.replace("</i>", "")
    outfile.write(data1)
```
