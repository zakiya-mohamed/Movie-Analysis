## Movie Colour Palettes
Usage of the K-Means Algorithm to determine colour palettes from movie stills 

- Reference: https://github.com/a-poor/color-palettes/blob/main/color-palettes.ipynb

```python
# Import required libraries
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

from PIL import Image
```

```python
# Change working directory
import os 
os.chdir('/Users/nmfamily/Desktop')
os.getcwd()
```

1) Process movie still and display

```python
# Specify path to movie still and display still
still_path = Path("Logan (2017).png")
still = Image.open(still_path.open("rb"))
still
```

2) Convert movie still to a numpy array and obtain specific dimensions

```python
arr = np.array(still)
arr.shape
```

3) Identify colour channels of movie still

```python
still.mode
```

3) Convert the 3D array (height, width, colour channel) into a 2D array (pixel-number, colour channel), obtain dimensions of 2D array

```python
pix = arr.reshape((-1,4))
pix.shape
```

4) Obtain range of values of pixels 

```python
print('Pixel minimum: {}'.format(pix.min()))
print('Pixel mean:    {}'.format(pix.mean()))
print('Pixel maximum: {}'.format(pix.max()))
```

5) Re-center pixels between the values 0 and 1

```python
pix = pix.astype("float32") / 255
print('Pixel minimum: {}'.format(pix.min()))
print('Pixel mean:    {}'.format(pix.mean()))
print('Pixel maximum: {}'.format(pix.max()))
```

6) Group pixels into a colour palette

```python
# Import required library
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 8)  # Specify K-Means clustering with 8 centres
preds = kmeans.fit_predict(pix)
centers = kmeans.cluster_centers_
```

```python
centers
```

```python
plt.figure(figsize = (14,8))
plt.imshow(still)
plt.grid()
plt.axis('off')
plt.show()

plt.figure(figsize = (14,8))
plt.imshow(centers[np.concatenate([[i] * 100 for i in range(len(centers))]).reshape((-1, 10)).T])
plt.title("Logan (2017) Color Palette")
plt.grid()
```

7) Apply the above algorithm to the rest of the movie stills

**(a) Transformers: The Last Knight (2017)**

```python
still_path = Path("Transformers - The Last Knight (2017).png")
still = Image.open(still_path.open("rb"))

arr = np.array(still)
pix = arr.reshape((-1,4))
pix = pix.astype("float32") / 255

kmeans = KMeans(n_clusters = 8) 
preds = kmeans.fit_predict(pix)
centers = kmeans.cluster_centers_

# Plot movie still
plt.figure(figsize = (14,8))
plt.imshow(still)
plt.grid()
plt.axis('off')
plt.show()
    
# Plot colour palette
plt.figure(figsize = (14,8))
plt.imshow(centers[np.concatenate([[i] * 100 for i in range(len(centers))]).reshape((-1,10)).T])
plt.title("Transformers: The Last Knight (2017) Color Palette")
plt.grid()
```
