## Movie Metadata Descriptive Statistics (Preliminary Analysis)

Metadata is collected from IMDb exclusively (apart from Metacritic user ratings which are collected from Metacritic, for the purpose of comparison). Crucial informations is visualized for all movies in order to identify any anomalies or similarities between movies over the years 2017 - 2019

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

1) Create DataFrame of movie metadata

```python
data = {'Movie':['Logan (2017)', 'Ready Player One (2018)', 'Ad Astra (2019)', 'Transformers: The Last Knight (2017)',
                 'Pacific Rim Uprising (2018)', 'Men In Black: International (2019)'], 
        'IMDb User Rating (/10)':[8.2, 7.4, 6.5, 5.2, 5.6, 5.6],
        'Gross Worldwide Box Office (100 Million USD)':[6.19179950, 5.82893671, 1.27461872, 6.05425157, 
                                                            2.90930148, 2.53890701],
       'Estimated Budget (100 Million USD)':[0.97000000, 1.75000000, 0.90000000, 2.17000000, 1.50000000,
                                               1.10000000],
       'Runtime (minutes)':[137, 140, 123, 154, 111, 114],
        'Director':['James Mangold', 'Steven Spielberg', 'James Grey', 'Michael Bay', 'Steven S. DeKnight', 
                    'F. Gary Gray'],
        'Metacritic User Rating (/10)':[8.4, 7.3, 6.2, 4.1, 4.8, 4.3],
        'Year of Release':[2017, 2018, 2019, 2017, 2018, 2019],
        'Number of IMDb Critic Reviews':[655, 602, 458, 298, 310, 275],
        'Number of IMDb User Reviews (1000)':[1.681, 2.137, 3.293, 0.917, 0.796, 1.186]}
```

2) Display Data Frame 

```python
df = pd.DataFrame(data)
df
```

3) Plot bar chart of movies against IMDB and Metacritic ratings

```python
fig = plt.figure(figsize = (15, 6))
fig.patch.set_facecolor('w')
ax = fig.add_axes([0,0,1,1])

X_axis = np.arange(len(df['Movie']))

plt.bar(X_axis - 0.2, df['IMDb User Rating (/10)'], 0.4, label = 'IMDb', color = 'gold')
plt.bar(X_axis + 0.2, df['Metacritic User Rating (/10)'], 0.4, label = 'Metacritic', color = 'royalblue')
  
plt.xticks(X_axis, df['Movie'])
plt.ylabel('User Rating (/10)')
plt.legend(fontsize = 12)
plt.show()
```

5) Plot bar chart of IMDb ratings against movies

```python
fig = plt.figure(figsize = (15, 6))
fig.patch.set_facecolor('w')
ax = fig.add_axes([0,0,1,1])

ax.bar(df['Movie'][:3], df['IMDb User Rating (/10)'][:3], color = 'mediumseagreen', label = 'Well-rated Movies')
ax.bar(df['Movie'][3:], df['IMDb User Rating (/10)'][3:], color = 'indianred', label = 'Poorly-rated Movies')

ax.set_ylabel('IMDb User Rating (/10)')
ax.legend(fontsize = 12)
plt.show()
```

6) Plot bar chart of movies against estimated budget

```python
fig = plt.figure(figsize = (15, 6))
fig.patch.set_facecolor('w')
ax = fig.add_axes([0,0,1,1])

ax.bar(df['Movie'][:3], df['Estimated Budget (100 Million USD)'][:3], color = 'mediumseagreen', label = 'Well-rated Movies')
ax.bar(df['Movie'][3:], df['Estimated Budget (100 Million USD)'][3:], color = 'indianred', label = 'Poorly-rated Movies')

ax.set_ylabel('Estimated Budget (100 Million USD)')
ax.legend(fontsize = 12)
plt.show()
```

7) Plot bar chart of movies against gross worldwide box office

```python
fig = plt.figure(figsize = (15, 6))
fig.patch.set_facecolor('w')
ax = fig.add_axes([0,0,1,1])

ax.bar(df['Movie'][:3], df['Gross Worldwide Box Office (100 Million USD)'][:3], color = 'mediumseagreen', label = 'Well-rated Movies')
ax.bar(df['Movie'][3:], df['Gross Worldwide Box Office (100 Million USD)'][3:], color = 'indianred', label = 'Poorly-rated Movies')

ax.set_ylabel('Gross Worldwide Box Office (100 Million USD)')
ax.legend(fontsize = 12)
plt.show()
```

8) Check for presence of correlation between variables in data frame using a heatmap

```python
fig = plt.figure(figsize = (8, 8))
fig.patch.set_facecolor('w')
sns.heatmap(df.corr())
plt.show()
```
