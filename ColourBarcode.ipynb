## Colour Barcode Analysis
Program that generates a 'colour barcode' for an input video file using the **K-means clustering** algorithm. One frame is read per second of the video and the frames **most dominant colour** is added as one strip in the colour barcode. This is repeated for all frames of the video until the video is over

### Overall Method
For this analysis in particular, since the video files are of large sizes and lengths, we choose to read the movie frames at **5 second intervals** to avoid computational difficulties. We specify the K-Means clustering with **10 clusters/centroids**
